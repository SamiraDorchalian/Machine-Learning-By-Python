SVM(Support Vector Machines)(Breast-Cancer)

Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Get the Data

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
cancer['feature_names']
```

Set Up DataFrame

```python
df_feat = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
df_feat.head()
```

```python
df_feat.info()
```

```python
df_feat.describe()
```

```python
df_target = pd.DataFrame(cancer['target'], columns=['Lable'])
df_target
```

```python
df_target.head()
```

Train Test Split

```python
X = df_feat
y = df_target['Lable']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
print(len(X_train))
print(len(X_test))
print('-----------------------------------------')
print(X_train.head())
print(X_test.head())
print('---------------------------------------------------------------------')
print(len(y_train))
print(len(y_test))
print('-----------------------------------------')
print(y_train.head())
print(y_test.head())
```

SVM Classifier

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

Evaluation

```python
pred = svc_model.predict(X_test)
pred
```

```python
df_eval = pd.DataFrame({'Lable': y_test, 'Prediction':pred})
df_eval
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```

Grid Search

```python
param_grid = {
    'C':[0.1, 1, 10, 100, 1000], 
    'gamma':[1, 0.1, 0.01, 0.001, 0.0001]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
pred = grid.predict(X_test)
pred
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC())
])
```

```python
param_grid = {
    'svm__C': [1, 10, 100, 1000],
    'svm__gamma': ['scale', 0.001, 0.0001],
    'svm__kernel': ['rbf']
}
```

```python
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=3)
grid = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy', verbose=3)
grid.fit(X_train, y_train)
```

```python
print("بهترین پارامترها:", grid.best_params_)
```

```python
print("دقت تست:", grid.score(X_test, y_test))
```
