فردی از سایت آمازون با پلاک t:90WT  خریدی انجام داده است . مشخص شده شماره پلاک اشتباه وارد شده و مبلغ باید به فرد بازگردد مبلغ این خرید را به دست آوردید

مبلغی پول از کارت یکی از مشتریان کسر شده که اضافی بوده و آمازون قصد دارد تا با ارسال ایمیلی به این کاربر مشکل پیش آمده رااطلاع دهد. ایمیل کاربری که شماره کارت او 4926535242672853 است را به دست اوردید.

```python
ecom[ecom['Credit Card'] == 4926535242672853]['Email']
```

شرکت امازون قصد دارد تا محبوب ترین دامین های ایمیلی مشترکین خود را بیابد و با عقد قرارداد با شرکت های صاحب این دامینها ، کاری کند که دیگر ایمیل های آمازون در این دامین ها به اسپم نرود. وظیفه شما این است که پنج هاست برتر دامین های ایمیل مشترکین را بیابد

```python
ecom['Email'].apply(lambda x:x.split('@')[1].split('.')[0]).value_counts().head()
```
