Logistic Regression

```python
X = adv[['Daily Time Spent on Site','Age','Area Income','Daily Internet Usage','Male']]
y = adv['Clicked on Ad']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logReg = LogisticRegression()
```

```python
logReg.fit(X_train, y_train)
```

Predictions and Evaluations

```python
pred = logReg.predict(X_test)
pred
```

```python
from sklearn import metrics
```

```python
metrics.confusion_matrix(y_test, pred)
```

```python
print(metrics.classification_report(y_test, pred))
```

```python
print(logReg.score(X_test, y_test)*100)
```

```python
logReg.predict([[80,27,52000,2500,0]])
```

```python
logReg.coef_
```

```python
Coefficients = pd.DataFrame(logReg.coef_.reshape(5,1), X.columns, columns=['Coeff'])
Coefficients
```
