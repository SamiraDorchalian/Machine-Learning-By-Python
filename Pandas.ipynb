```python
import numpy as np
import pandas as pd
```

# Series

```python
labels = ['a','b','c']
my_list = [10,20,30]
arr = np.array([10,20,30])
dic = {'a': 10, 'b': 20, 'c': 30}
```

# Creating Series

## using Lists

```python
pd.Series(my_list)
```

```python
a = pd.Series(my_list)
a
```

```python
a[0]
```

```python
b = pd.Series(data=my_list, index=labels)
b
```

## using Array

```python
pd.Series(arr)
```

```python
pd.Series(data=arr, index=labels)
```

## using Dectionary

```python
pd.Series(dic)
```

```python
pd.Series(data=dic, index=labels)
```

## using an Index

```python
ser_1 = pd.Series(data=[1,2,3,4], index=['Iran', 'Korean', 'China', 'Japan'])
ser_1
```

```python
ser_1['Iran']
```

```python
ser_1['Japan']
```

```python
ser_1 = pd.Series(data=[1,2,3,4], index=['Iran', 'Korean', 'China', 'Japan'])
ser_2 = pd.Series(data=[5,7,9,11], index=['Iran', 'korean', 'USA', 'Japan'])
```

```python
ser_1
```

```python
ser_2
```

```python
ser_1 + ser_2
```

```python
ser_1 - ser_2
```
