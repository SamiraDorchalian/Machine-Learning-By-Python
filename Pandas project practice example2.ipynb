<div style="direction:rtl">
یک بنگاه کاریابی قصد دارد که مشاغلی که سال 2013 در آن ها فقط یک نفر مشغول کار بودند را شناسایی کند و تعدادی کارمند به آن مشاغل معرفی کند. وظیفه ی شما این است که تعداد این مشاغل را به دست آورید.

```python
sal_2013 = sal[sal['Year'] == 2013]
job_counts_2013 = sal_2013['JobTitle'].value_counts()
single_person_jobs = job_counts_2013[job_counts_2013 == 1]
num_single_jobs = len(single_person_jobs)
num_single_jobs
```

<div style="direction:rtl">
یک بنگاه کاریابی به شما پروژه ای سفارش کرده که در آن باید تمام مشاغلی که در آنها کلمه ی chief وجود دارد را پیدا کنید. تابعی برای این کار بنویسید و این تعداد را به دست آورید.

```python
def chief_string(title):
    return 'chief' in str(title).lower()
    
chief_jobs_count = sal['JobTitle'].apply(chief_string).sum()
chief_jobs_count
```

سپس ستونی تحت عنوان طول نام شغل ایجاد کنید و درباره ی همبستگی احتمالی طول نام شغل و میزان مجموع دریافتی جدولی را ارئه دهید.

[correlation in pandas](https://stackoverflow.com/questions/42579908/use-corr-to-get-the-correlation-between-two-columns)

```python
import numpy as np
sal['JobTitleLength'] = sal['JobTitle'].apply(lambda x: len(str(x)))

correlation = sal['JobTitleLength'].corr(sal['TotalPayBenefits'])

correlation
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(data=sal, x='JobTitleLength', y='TotalPayBenefits', alpha=0.5)
plt.title('(TotalPayBenefits)')
plt.show()
```
